```python
chg = data.pct_change().dropna()
### describe the schema of the data and metrics
chg.describe()
```

```python
### To transpose the data of rows (stocks) and mean,std we use `T`
total = chg.describe().T.loc[:,['mean','std']]
### the mean values refered with all days
total
```

### if below cell is re-executed again, re-run the above cell so the multi factor doesn't change the graph

```python
### to make the mean value for year - 252 tranding days in an year
total['mean'] = total['mean']*252
total['std'] = total['std'] * numpy.sqrt(252)
total
```

```python
## plot
total.plot.scatter(x='std', y='mean',figsize=(17,8), fontsize=18)
for i in total.index:
    ## this is to annotate the dots plotted, and the +0.002 indicates how far
    ## the annotation to be displayed
    plt.annotate(i, xy=(total.loc[i,'std']+0.002,total.loc[i,'mean']+0.002),size = 15)
```

### from the graph above the further away the mean value, the high risk, high returns over period
#### AMZN - has high returns and also high risk
#### INTC - has low returns and low risk for a period of time
